# 10 — QSNN Training with Parameter-Shift Rule

Train a quantum spiking neural network layer using the parameter-shift gradient rule.
Each CRy synapse angle is updated via $(L(\theta+\pi/2) - L(\theta-\pi/2))/2$.

```python
import matplotlib.pyplot as plt
import numpy as np

from scpn_quantum_control.qsnn.qlayer import QuantumDenseLayer
from scpn_quantum_control.qsnn.training import QSNNTrainer
```

## 1. Build Layer and Trainer

```python
layer = QuantumDenseLayer(n_neurons=2, n_inputs=3, seed=42)
trainer = QSNNTrainer(layer, lr=0.05)
print(f"Layer: {layer.n_inputs} inputs, {layer.n_neurons} neurons")
print(f"Initial weights:\n{layer.get_weights()}")
```

## 2. Generate Training Data

```python
rng = np.random.default_rng(42)
X = rng.uniform(0, 1, (20, 3))
y = np.column_stack(
    [
        (X[:, 0] + X[:, 1]) / 2,
        (X[:, 1] + X[:, 2]) / 2,
    ]
).clip(0, 1)
print(f"Training: {X.shape[0]} samples")
print(f"Input range: [{X.min():.2f}, {X.max():.2f}]")
print(f"Target range: [{y.min():.2f}, {y.max():.2f}]")
```

## 3. Train

```python
losses = trainer.train(X, y, epochs=10)

fig, ax = plt.subplots(figsize=(7, 4))
ax.plot(range(1, len(losses) + 1), losses, "o-", color="#1f77b4", linewidth=2)
ax.set_xlabel("Epoch")
ax.set_ylabel("Mean MSE Loss")
ax.set_title("QSNN Training: Parameter-Shift Gradient Descent")
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"Loss: {losses[0]:.4f} -> {losses[-1]:.4f}")
print(f"Final weights:\n{layer.get_weights()}")
```
